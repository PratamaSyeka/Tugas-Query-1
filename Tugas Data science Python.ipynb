```
import pandas as pd
import os # Import the os module for directory creation
import shutil # Import shutil for file operations (optional, but good practice)

source_file = 'data penggunaan listrik.csv'
destination_dir = '/mnt/data/'
destination_file = os.path.join(destination_dir, source_file)

!mkdir -p {destination_dir}


if not os.path.exists(source_file):
    print(f"Error: File sumber '{source_file}' tidak ditemukan di direktori saat ini.")
    print("Pastikan file 'data penggunaan listrik.csv' sudah diunggah atau berada di direktori yang benar.")

else:

    try:
        shutil.copy(source_file, destination_file)
        print(f"File '{source_file}' berhasil disalin ke '{destination_dir}'.")
    except Exception as e:
        print(f"Error saat menyalin file: {e}")
        # Handle error copy jika terjadi

    # Periksa kembali apakah file berhasil dicopy ke direktori tujuan
    if os.path.exists(destination_file):
        # Membaca file yang diunggah untuk melihat strukturnya
        file_path = destination_file # Gunakan path tempat file disalin
        try:
            data_listrik = pd.read_csv(file_path)

            # Menampilkan beberapa baris pertama untuk memahami strukturnya
            print("Data berhasil dimuat. Berikut adalah beberapa baris pertama:")
            display(data_listrik.head()) # Menggunakan display untuk output yang lebih rapi di notebook

        except FileNotFoundError:
             print(f"Error: File '{file_path}' masih tidak ditemukan setelah disalin.")
        except Exception as e:
            print(f"Error saat membaca file CSV: {e}")

    else:
        print(f"Error: File '{destination_file}' tidak ditemukan di direktori tujuan setelah mencoba menyalin.")

```
